```python
import pandas as pd

df = pd.read_csv('roto-point.csv', index_col=0)
df

```

```python
import numpy as np

# green background
style_top = 'background-color:#e6ffe6'

# red background
style_bottom = 'background-color:#ffe6e6'

def highlight_max_min(s):
    """
    Highlight the maximum value in a Series with a green background
    and the minimum value with a red background.
    
    Parameters
    ----------
    s : pandas.Series
        The Series to be styled.
    
    Returns
    -------
    pandas.Series
        The styled Series with the maximum and minimum values highlighted.
    """
    is_max = s == s.max()
    is_min = s == s.min()
    return [style_top if v else style_bottom if m else '' for v, m in zip(is_max, is_min)]

def remove_trailing_zeros(x):
    """
    Remove trailing zeros from a number.
    
    Parameters
    ----------
    x : float
        The number to be formatted.
    
    Returns
    -------
    str
        The formatted number as a string without trailing zeros.
    """
    return ('%f' % x).rstrip('0').rstrip('.') if isinstance(x, float) else x

    
# Apply the highlight_min function to each column
styled_df = df.style.apply(highlight_max_min, axis=0).format(remove_trailing_zeros).set_table_styles(
        [{'selector': 'thead th', 'props': [('background-color', '#f7f7f9'), ('color', '#333333')]}]
    ).set_properties(**{'text-align': 'right', 'width': '50px'}).set_caption("Styled DataFrame")

# Display the styled DataFrame
styled_df
```

```python
# Save styled DataFrame as HTML
html = styled_df.to_html(col_space=60, na_rep='', bold_rows=False, notebook=True)
print(html)
```
